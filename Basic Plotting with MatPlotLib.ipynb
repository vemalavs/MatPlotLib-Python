```python
#Script1
import matplotlib.pyplot as plt
import numpy as np
import math

x_vals = np.linspace(0,20,20)#20 equidistant numbers generated
y_vals = [math.sqrt(i) for i in x_vals]
plt.plot(x_vals, y_vals)
```

```python
#Script3 #change plot size
import matplotlib.pyplot as plt
import numpy as np
import math

plt.rcParams ["figure.figsize"] = [8,6]#8 in wide 6in tall
#figure.figsize attribute

x_vals = np.linspace (0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.plot(x_vals, y_vals)
```

```python
#Script4 #adding string labels
import matplotlib.pyplot as plt
import numpy as np
import math

x_vals = np.linspace(0,20,20)
y_vals =[math.sqrt(i) for i in x_vals]
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Square Roots')
plt.plot(x_vals, y_vals)
```

```python
#Script5 #Specify color of line plot
import matplotlib.pyplot as plt
import numpy as np
import math

x_vals = np.linspace(0,20,20)
y_vals =[math.sqrt(i) for i in x_vals]
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title ('Square Roots')
plt.plot(x_vals, y_vals, 'r')# r=red
```

```python
#Script6 #Add a legend
import matplotlib.pyplot as plt
import numpy as np
import math

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Square Roots')
plt.plot(x_vals, y_vals, 'r', label ='Square Root')
plt.legend(loc = 'upper center')
#label attribute of legend, n loc to specify the location
```

```python
#Script7 #Plot Multiple Lines
import matplotlib.pyplot as plt
import numpy as np
import math

x_vals = np.linspace(0,20,20)
y_vals = [math.sqrt(i) for i in x_vals]
y2_vals = x_vals **3 #** exponentiation 
plt.xlabel ('X Values')
plt.ylabel('Y Values')
plt.title('Square Roots')
plt.plot(x_vals, y_vals, 'r', label ='Square Root')
plt.plot(x_vals, y2_vals, 'b', label = 'Cube')
plt.legend(loc = 'upper center')
```
